```python
import numpy as np # For working with arrays
import matplotlib.pyplot as plt # For plotting
import pandas as pd # For importing csv data
import sympy as sp # For symbolic differentiation
import math
```

```python
def orderedpairs(filename):
  import numpy as np
  import pandas as pd

  data = pd.read_csv(filename)
  X = data['X']
  Y = data['Y']

  terrain = ''
  i=0
  while i< len(X):
    terrain += '{'
    terrain += f'{X[i]},{Y[i]}'
    terrain += '}'
    if i< (len(X)-1):
      terrain += ','
    i += 1
  return(terrain)
```

```python
orderedpairs('terrain.csv')
```
